### Visualisasi Data
- 3 tipe visualisasi
- pie chart
- simpul cell markdown
- gunakan Title, xlabel, ylabel

```python
import pandas as pd
from matplotlib import pyplot as plt


# untuk membaca file csv
df = pd.read_csv('sets.csv')
df
```

#### Pie Chart

```python
df = df.dropna(subset=['year'])
pie = df['year'].value_counts().rename_axis('year').reset_index(name='tahun')

plt.figure(figsize=(10,10))
plt.pie(pie.tahun, labels=pie.year, startangle=2, autopct='%.1f%%')
plt.title('year')
plt.show()
```

##### menampilkan rentan perbandingan persentase jumlah part di setiap tahun pada grafik pie

#### Grafik Plot dan Garis

```python
plt.plot(df.year, df.num_parts)
```

##### pada grafik diatas menampilkan nomor parts dari tahun 1950 dampai 2020

#### Grafik Batang

```python
y = list(df.iloc[:,2])
x = list(df.iloc[:,2])

plt.bar(x, y, color='blue')
plt.title("sets")
plt.xlabel("year")
plt.ylabel("number parts")
```

##### menampilkan year dan num_parts  dengan xlabel = year dan ylabel = number parts
